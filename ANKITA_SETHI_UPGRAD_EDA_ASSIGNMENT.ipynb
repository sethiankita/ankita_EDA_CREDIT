### Creating a data frame for the null values in application_data having null values more than or equal to 40 % according to me if more than 40% data is missing then its of no use but we can't delete that data so i am creating a seperate dataframe for that.

```python
missing_val=pd.DataFrame(app_data.isnull().sum().sort_values()).reset_index()
missing_val.rename(columns={'index':'Name',0:'nullvalue_count'},inplace=True)
#displaying percentage of missing values from application_Data
missing_val["percent_miss_val"]=missing_val["nullvalue_count"]/app_data.shape[0]*100
missing_val
```

## Dropping 
    40% values because if more than 40% values are missing then it usually indicates the value is of lesser use.   

```python
removed=missing_val[missing_val["percent_miss_val"]>=40]
removed[removed["percent_miss_val"]>=40]['Name']
x=removed[removed["percent_miss_val"]>=40]['Name'].to_list()
x
```

```python
#Dropping the columns having missing value more than 40%
app_miss_val_removed=app_data.drop(x,axis=1)
app_miss_val_removed.head()
```

## Treating missing value from different column 
    

### 1.Occupation Type
    Since occupation type is a categorical data and occupation column is important column for giving loan so we
    will not drop this column. the safest way to treat missing value in 
    categorical variable is mode if it would have been numerical variable than mean and median is safer thing to do.

```python
flag_drop_df.OCCUPATION_TYPE.value_counts().sort_values()
```

```python
#Calculating Mode
flag_drop_df.OCCUPATION_TYPE.mode()[0]
```

```python
#filling missing data with mode value and then checking null value 
flag_drop_df.OCCUPATION_TYPE=flag_drop_df.OCCUPATION_TYPE.fillna(flag_drop_df.OCCUPATION_TYPE.mode()[0])
flag_drop_df.OCCUPATION_TYPE.isnull().sum()
```

### 2. NAME_TYPE_SUITE
    Since the value of unaccompanied is highest that is in lakhs as compared to the other values we will replace by the mode values.

```python
flag_drop_df.NAME_TYPE_SUITE.value_counts()
```

```python
flag_drop_df["NAME_TYPE_SUITE"]=flag_drop_df["NAME_TYPE_SUITE"].fillna(flag_drop_df["NAME_TYPE_SUITE"].mode()[0])
flag_drop_df["NAME_TYPE_SUITE"].isnull().sum()
```

### 3.  CODE_GENDER
